```python
from pathlib import Path
import sys

from dotenv import load_dotenv

def find_project_root():
    for path in (Path.cwd(), *Path.cwd().parents):
        if (path / "minimax_model.py").exists():
            return path
    raise FileNotFoundError("Could not find minimax_model.py")

PROJECT_ROOT = find_project_root()
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

load_dotenv(PROJECT_ROOT / ".env", override=True)
```

## No memory

```python
from langchain.agents import create_agent
from minimax_model import get_model


agent = create_agent(
    get_model()
)
```

```python
from langchain.messages import HumanMessage

question = HumanMessage(content="Hello my name is Seán and my favourite colour is green")

response = agent.invoke(
    {"messages": [question]} 
)
```

```python
from pprint import pprint

pprint(response)
```

```python
question = HumanMessage(content="What's my favourite colour?")

response = agent.invoke(
    {"messages": [question]} 
)

pprint(response)
```

## Memory

```python
from minimax_model import get_model
from langgraph.checkpoint.memory import InMemorySaver  


agent = create_agent(
    get_model(),
    checkpointer=InMemorySaver(),  
)
```

```python
from langchain.messages import HumanMessage

question = HumanMessage(content="Hello my name is Seán and my favourite colour is green")
config = {"configurable": {"thread_id": "1"}}

response = agent.invoke(
    {"messages": [question]},
    config,  
)
```

```python
pprint(response)
```

```python
question = HumanMessage(content="What's my favourite colour?")

response = agent.invoke(
    {"messages": [question]},
    config,  
)

pprint(response)
```
